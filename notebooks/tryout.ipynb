```python
def load_data(email, password):
    info = pd.read_excel("data/Werknemers_gegevens - Test.xlsx", index_col=0, sheet_name= "TraineesMaria")
    matches = info.loc[info["Emailadres"] == email, ["Wachtwoord", "Persoons_ID"]]
    user_id = matches["Persoons_ID"].iloc[0]
    print(user_id)
    if not matches.empty and matches["Wachtwoord"].iloc[0] == password: 
        try:
            df_peer = pd.read_excel("data/employee_name.xlsx", sheet_name="Peer Reviews", index_col=0)
            df_client = pd.read_excel("data/employee_name.xlsx", sheet_name="Opdrachtgever Reviews", index_col=0)
            df_peer.index = pd.to_datetime(df_peer.index)
            df_client.index = pd.to_datetime(df_client.index)   
            return df_peer, df_client
        except Exception as e:
            print(f"Fout bij het laden van data voor {user_id}: {e}")
            return None, None
    else:
        print("Dit e-mailadres en/of wachtwoord zijn incorrect. Ga naar de accountmanager als je je wachtwoord bent vergeten.")

load_data(email, password)
```
